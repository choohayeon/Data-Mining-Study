#### 리스트, ndarray를 DataFrame으로 변환
- 1차원

```python
import numpy as np

col_name1=['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

#리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

#넘파이를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

- 2차원 

```python
# 3개의 칼럼명이 필요함.
col_name2=['col1', 'col2', 'col3']

# 2행X3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1,2,3],
          [11,12,13]]
array2 = np.array(list2)
print('array2 shape:',array2.shape) 
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

#### 딕셔너리를 DataFrame으로 변환
- 딕셔너리 키(Key)은  컬럼명(문자열로 구성)으로 매핑
- 딕셔너리 값(Value)은 해당하는 칼럼 데이터(리스트 또는 ndarray로 구성)로 매핑

```python
dict= {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

- <span style="color:black; font-size:1.3em"> DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환</span>

#### DataFrame를 ndarray로 변환

```python
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:',array3.shape)
print(array3)
```

#### DataFrame를 리스트, 딕셔너리로 변환
- 리스트로의 변환: values로 얻은 ndarray에 tolist() 호출
- 딕셔너리로의 변환: DataFrame 객체의 to_dict() 메서드 호출, 인자:'list'입력 시, 딕셔너리의 값->리스트형으로 반환됨

```python
#DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

#DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### 4. DataFrame 데이터 삭제
- drop()
> <span style="color:black; font-size:1.4em"> DataFrame.drop( <span style="color:red">labels=None, axis=0</span>, index=None, columns = None, level = None, <span style="color:red">inplace=False</span>, error='raise')</span>
>> - <span style="color:black; font-size:1.3em">  labels: 드롭하길 원하는 칼럼명 입력</span>
>> - <span style="color:black; font-size:1.3em">  axis: 값에 따라 특정 칼럼 또는 특정 행을 드롭(axis 0: 로우 방향/ axis 1: 칼럼 방향)</span>
>> - <span style="color:black; font-size:1.3em">  inplace: </span>
>>> -  Inplace = False(디폴트 값) 즉, 자기 자신의 DataFrame의 데이터는 삭제하지 않으며, 삭제된 DataFrame을 반환
>>> -  Inplace = True 는 자신의 DataFrame의 데이터 삭제


#### titanic_df ['Age_0']열 삭제

### 데이터 셀렉션 및 필터링
- DataFrame 바로 뒤의 [ ] 연산자는 넘파이의 [ ]나 Series의 [ ]와 다르다.
- DataFrame 바로 뒤의 [ ] 내의 입력값은 칼럼명(또는 칼럼의 리스트)을 지정해 칼럼 지정 연산에 사용하거나 불린 인덱스 용도로만 사용해야한다.
- DataFrame[0:2]와 같은 슬라이싱 연산으로 데이터를 추출하는 방법은 사용하지 않는게 좋습니다.
> - 판다스 [ ] 연산자: 칼럼명 문자(칼럼명의 리스트 객체), 인덱스로 변환 가능한 표현식 즉, 칼럼 지정 연산자
> - 넘파이 [ ] 연산자: 행의 위치, 열의 위치, 슬라이싱 범위등을 지정해 데이터 가져오기

### 정렬 Aggregation 함수, Groupby 적용
- DataFrame, Series 정렬: sort_values()
    > - RDBMS의 order by 키워드와 유사
    > - main parameter(by, ascending, inplace)
    >> - by: 특정 칼럼을 입력하면 해당 칼럼으로 정렬 수행
    >> - ascending: 오름차순, 내림차순 정렬, 디폴트 값은 ascending=True(오름차순)/ ascending=False(내림차순)
    >> - inplace: 원본 DataFrame 그대로 유지할 것인지, 정렬을 적용할 것인지 결정, 디폴트 값은 inplace=False(그대로 유지, 반환 값 존재)/ inplace=True(원본에 정렬 적용, 반환값:None)

```python
titanic_sorted = titanic_df.sort_values(by=['Name'])
titanic_sorted.head(3)
```

#### 여러 개의 칼럼으로 정렬 - by에 리스트 형식의 정렬하는 칼럼 입력

```python
titanic_sorted = titanic_df.sort_values(by=['Pclass','Name'],ascending = False)
titanic_sorted
```

#### groupby()적용
- DataFrame.grouby(by='col') : 해당 칼럼으로 groupby
- sort_values()와 다른점<span style="color:red; font-size:1.3em">
> - DF.sort_values(by=['컬럼명'])
> - DF.groupby(by='컬럼명') </span> 
  

```python
titanic_groupby = titanic_df.groupby(by='Pclass')
print(type(titanic_groupby))
```

```python
titanic_groupby = titanic_df.groupby(by='Pclass').count()
titanic_groupby
```

>> DataFrame의 groupby()에 특정 칼럼만 aggregation함수 적용:
>> - groupby()로 반환된 DataFrameGroupby 객체에 해당 칼럼을 필터링 -> aggregation 함수 적용

```python
titanic_grouby = titanic_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
titanic_grouby
```

```python
titanic_df.groupby(by='Pclass')['Age'].agg([max,min])
```

#### 여러 개의 칼럼이 서로 다른 aggregation 함수를 groupby에서 호출하는 경우: 딕셔너리 형태 이용

```python
agg_format={'Age':'max', 'SibSp':'sum','Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```
